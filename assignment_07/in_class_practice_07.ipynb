```python
import pandas as pd
print("Pandas version is:",pd.__version__)

import plotly
print(plotly.__version__)
import plotly.graph_objects as go
import plotly.express as px

```

```python
df = pd.read_csv("/content/wdi_data (1).csv")
df.head(10)
```

```python
#df.drop(columns=["unnamed:0"], inplace=True)
df_new =df.drop(columns=["Unnamed: 0"])
df_new.head()
```

```python
df.head()
```

```python
#df_ind = df_new.query(""Country Code'"=="IND") #not working becauase of space btw the column name
df_ind = df_new[df_new["Country Code"]=="IND"]
df_ind
```

```python
df_ind_gdpcap = df_ind.query("indicator == 'NY.GDP.PCAP.PP.CD'	")
df_ind_gdpcap
```

```python
df_ind_gdpcap.info()
```

```python
fig = go.Figure()

trace_0 = go.Bar(
    
    x=df_ind_gdpcap['Year'],
    y=df_ind_gdpcap['value']
)
fig.add_trace(trace_0)


fig.update_layout(
    title = "India GDP per captia over time ",
    xaxis_title='Year',
    yaxis_title ='GDP per capita'
)

fig.show()
```

```python
fig = px.bar(
    data_frame = df_ind_gdpcap,
    title="India GDP per capita over time",
    labels={"value":"GDP per capita","Year":"YEAR"},
    x='Year',
    y='value'
)
fig.show()
```
